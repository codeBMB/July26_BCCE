## Section 2 — Runtime Type

A **runtime** is the computing environment that actually runs your code. Colab gives you a choice of hardware and coding language (Python, R, or Julia).

The type of hardware you need will vary. For small datasets (fewer than 100,000 rows), fitting enzyme kinetics, or visualizing most BMB-related plots, a **CPU** (Central Processing Unit) runtime is fine. For molecular dynamics or protein structure modeling, a **GPU** (Graphics Processing Unit) runtime might be needed for reasonable speed.

To change your runtime, go to **Runtime → Change runtime type** and select the hardware accelerator you need. Start with CPU unless a specific tutorial tells you otherwise — GPU/TPU runtimes are a shared, limited resource.

## Section 7 — Debugging Practice: The Uninitialized List Bug

This is one of the most common early Python errors, so it's worth seeing on purpose. Try running the cell below. See if you can figure out why it doesn't work *before* reading the explanation underneath.

````python
Km, Vmax = 15.0, 100.0
substrate_concs = [1.0, 2.0, 3.0, 4.0, 6.0, 8.0, 10.0, 15.0, 20.0, 30.0, 50.0, 75.0, 100.0]

for number in substrate_concs:
    V = Vmax * number / (Km + number)
    velocities.append(V)

print(velocities)
````

This code doesn't work because on the first iteration of the loop, the list `velocities` doesn't exist yet. To fix it, we have to create the list *before* the loop starts. It can be empty when we start it, but it has to exist.

## Section 9 — More Practice with Logic

You can combine multiple conditions in an `if` statement using `and`, `or`, and `not`. Here's an example that keeps values at or below `Km`.

````python
Km, Vmax = 15.0, 100.0
substrate_concs = [1.0, 2.0, 3.0, 4.0, 6.0, 8.0, 10.0, 15.0, 20.0, 30.0, 50.0, 75.0, 100.0]
v_at_or_below_km = []

for number in substrate_concs:
    if number < Km or number == Km:
        velocity = Vmax * number / (Km + number)
        v_at_or_below_km.append(velocity)

print(v_at_or_below_km)
````
