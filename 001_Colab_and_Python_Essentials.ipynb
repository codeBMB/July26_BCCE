## Section 8 — Variables and Data Types

Any Python interpreter can work just like a calculator. This is not very useful on its own — try running the following cell. What do you think the output will be?

Here, Python has performed a calculation for us. To save this value, or other values, we assign them to a variable for later use. The syntax for assigning variables is the following:

```python
variable_name = variable_value
```

Let's put this to work with something a biochemist would actually calculate: the **Michaelis-Menten equation**, which describes enzyme reaction velocity:

$$v = \frac{V_{max} \cdot [S]}{K_m + [S]}$$

In the next cell, declare `Km = 15.0`, `Vmax = 100.0`, and a substrate concentration `substrate_conc = 50.0` (micromolar), then calculate and print `velocity` using the equation above.

````python
# Calculations using the Michaelis-Menten Equation
````

<details>
<summary>🔍 CLICK TO SEE ANSWER</summary>

```python
# 1. Declare the parameters
Km = 15.0     # micromolar
Vmax = 100.0  # nanomoles/second
substrate_conc = 50.0  # micromolar

# 2. Set up the equation
velocity = Vmax * substrate_conc / (Km + substrate_conc)

# 3. Print the results
print(velocity)
```
Note: You can change the values and see what happens!
</details>

Notice several things about this code. You can use `#` to add comments to your code, both at the start of a line and in the middle of the line — the computer does not act on these comments. They're used here to remind the user what units are used for each value.

When choosing variable names, you should pick something descriptive (`velocity` rather than `v` or `x`) — your future self will thank you.

````python
print(velocity)
````

Note that if you do not specify a new name for a variable, then referencing it doesn't automatically change the value of that variable. However, you *can* reassign a variable to a new value by overwriting it — this is worth doing carefully.

````python
print(velocity)
velocity = velocity * 60
print(velocity)
````

There are situations where it is reasonable to overwrite a variable with a new value, but you should always think carefully about this. Usually it is better practice to give the variable a new name and leave the existing variable as-is.

````python
# change velocity back
velocity = velocity / 60
v_nmols_per_min = velocity * 60
print(velocity)
print(v_nmols_per_min)
````

## Section 10 — Lists

Another common data structure in Python is the **list**. Lists group several values together and are declared using square brackets `[ ]`. **Python assigns special meanings to square brackets `[]`, parentheses `()`, and curly brackets `{}`, so you must be very careful with these characters.** List values are separated by commas.

The built-in function `len()` can be used to determine the length of a list.

````python
# This is a list
substrate_concs = [1.0, 2.0, 3.0, 4.0, 6.0, 8.0, 10.0, 15.0, 20.0, 30.0, 50.0, 75.0, 100.0]  # micromolar

# I can determine its length
s_length = len(substrate_concs)

# Print the length of the list
print('This list contains', s_length, 'substrate concentrations')
````

Note the format of the print statement above:

```python
print('text', variable, 'text')
```

f-strings are another approach to printing in Python, and they're simpler to format — you never have to guess where the spaces will be:

```python
print(f'text {variable} text {variable}')
```

Try this print statement:

````python
print(f'This list contains {s_length} substrate concentrations.')
````

You can use an element of a list as a variable in a calculation by referencing its position, or **index**. Remember that counting starts at zero, so the first element of a list is `list_name[0]`.

````python
# Convert the last substrate concentration to nM
concentration_nM = substrate_concs[12] * 1000
print(concentration_nM)
````

You can also add something to the end of a list using the `.append()` function.

````python
substrate_concs.append(1000.0)
print(substrate_concs)
````

## Section 11 — Slices

Sometimes you will want to make a new list that is a subset of an existing list — for example, just the first few elements. This is called a `slice`. The general syntax is

```python
new_list = list_name[start:end]
```

If you do not include a start index, the slice automatically starts at `list_name[0]`. If you do not include an end index, the slice automatically goes to the end of the list. Remember that the end index is **not included** in the result.

````python
# Reset the substrate_concs list
substrate_concs = [1.0, 2.0, 3.0, 4.0, 6.0, 8.0, 10.0, 15.0, 20.0, 30.0, 50.0, 75.0, 100.0]  # micromolar
short_list = substrate_concs[0:3]
print(short_list)
````

## Section 12 — For Loops: Repeating an Operation Many Times

Often, you will want to do something to every element of a list. The structure to do this is called a `for` loop. The general structure is:

```python
for variable in list:
    do things using variable
```

Indentation is very important in Python. There is nothing like an `end` statement that tells you the loop is finished — the indentation shows you what statements belong to the loop.

````python
substrate_concs = [1.0, 2.0, 3.0, 4.0, 6.0, 8.0, 10.0, 15.0, 20.0, 30.0, 50.0, 75.0, 100.0]
Km, Vmax = 15.0, 100.0

for number in substrate_concs:
    velocity = Vmax * number / (Km + number)
    print(velocity)
````

It would be even better if, instead of just printing the values, we saved them in a new list. To do this, we use the `append` function — but the list has to already exist (even if empty) *before* the loop starts.

````python
velocities = []
substrate_concs = [1.0, 2.0, 3.0, 4.0, 6.0, 8.0, 10.0, 15.0, 20.0, 30.0, 50.0, 75.0, 100.0]

for number in substrate_concs:
    V = Vmax * number / (Km + number)
    velocities.append(V)

print(velocities)
````

## Section 13 — Making Choices: Logic Statements

Within your code, you may need to evaluate a variable and then do something if it has a particular value. This type of logic is handled by an `if` statement.

````python
temp = 35       # a number we want to check
if temp > 30:   # a condition ends with a colon
  print("Hot")  # the action is indented
````

You can combine an `if` statement with a `for` loop — for example, when we're only interested in the lower substrate concentrations, as we might do when looking for the initial linear portion of a Michaelis-Menten curve.

````python
Km, Vmax = 15.0, 100.0
linear_MM = []

substrate_concs = [1.0, 2.0, 3.0, 4.0, 6.0, 8.0, 10.0, 15.0, 20.0, 30.0, 50.0, 75.0, 100.0]
for number in substrate_concs:
    if number < Km/4:
        V_linear = Vmax * number / (Km + number)
        linear_MM.append(V_linear)

print(linear_MM)
````

Other logic operations include:
- equal to `==`
- not equal to `!=`
- greater than `>`
- less than `<`
- greater than or equal to `>=`
- less than or equal to `<=`

You can also use `and`, `or`, and `not` to check more than one condition at a time — we'll practice this more in the extra practice notebook.
